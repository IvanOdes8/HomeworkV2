```python
visit = pd.read_csv(r'D:\Netology\Pandas\доп_материалы_к_пандас_1_и_2\visit_log.csv' ,delimiter=r";")
visit.head()
```

```python
# функция для столбца source_type с правилами
def rules (row) :
    if row ['traffic_source'] == ('yandex' or 'google') :
        return "organic"
    if row ['region'] == 'Russia' and row ['traffic_source'] == ('paid' or 'email') :
        return "ad"
    if row ['region'] != 'Russia' and row ['traffic_source'] == ('paid' or 'email') :
        return "other"
    return row ['traffic_source']
```

```python
# применение функции
visit['source_type'] = visit.apply(rules, axis = 1)
visit.head(20)
```

```python
ratings = pd.read_csv(r'D:\Netology\Pandas\ml-latest-small\ratings.csv')
ratings.head()
```

```python
ratings['date'] = pd.to_datetime(ratings['timestamp'], unit = 's')
ratings.head()
```

```python
ratings['dif_days'] = ratings.groupby('userId')['date'].transform(lambda x: (x.max()-x.min()).days)
ratings[ratings['userId']==4].head()
```

```python
# проверочная df для min даты
min_data = ratings.groupby(['userId'], as_index=False)['date'].min()
min_data.head(12)
```

```python
# проверочная df для max даты
max_data = ratings.groupby(['userId'], as_index=False)['date'].max()
max_data.head(12)
```

```python
# df только свыше 100 оценок
count_userID = ratings.groupby(['userId']).count()
userID_100 = count_userID.loc[count_userID['movieId'] >100]
userID_100.head()
```

```python
# merge для userID свыше 100 оценок c временем жизни
userID_100_lifetime = userID_100.merge(ratings[['userId','dif_days']], on = 'userId', how = 'left')
userID_100_lifetime.head()
```

```python
# среднее время жизни userID свыше 100 оценок
userID_100_lifetime['dif_days_y'].mean()
```
