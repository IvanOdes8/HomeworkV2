```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]

amount_word = {}
amount_in_queries = len(queries)

for query in queries:
    words = len(query.split(' '))
    if words in amount_word:
        amount_word[words] += 1
    else:
        amount_word[words] = 1

for word in amount_word:
    print(f'Поисковых запросов, содержащих {word} слов(а): {amount_word[word]/amount_in_queries:.2%}')
```

```python
results = {
    'vk': {'revenue': 103, 'cost': 98},
    'yandex': {'revenue': 179, 'cost': 153},
    'facebook': {'revenue': 103, 'cost': 110},
    'adwords': {'revenue': 35, 'cost': 34},
    'twitter': {'revenue': 11, 'cost': 24},
}

for i in results:
    results[i]['ROI'] = round((results[i]['revenue']/results[i]['cost']-1)*100, 2)
    New_results = sorted(results.items())
for i in New_results:
    print(i)
```
