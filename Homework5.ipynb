```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}

# Функция поиска владельца документа по его номеру
def search_name_by_code():
    if Input_command == 'p':
        Input_code = input('Введите код: ')
        def search_name_by_code_sub(Input_code, name_by_cod):
            for document_in_list in documents:
                if document_in_list['number'] == Input_code:
                    name_by_cod = document_in_list['name']
                    return name_by_cod
        if search_name_by_code_sub(Input_code, 'name') is None:
            print('Документ отсутствует')
        else:
            print('Владелец документа:', search_name_by_code_sub(Input_code, 'name'))
            

# Подфункция поиска полки на которой хранится документ по его номеру
def search_directory_by_cod_sub(Input_code):
    for num_of_directory in directories:
        if Input_code in directories[num_of_directory]:
            return (num_of_directory)   
# Функция поиска полки на которой хранится документ по его номеру
def search_directory_by_cod():
    if Input_command == 's':
        Input_code = input('Введите код: ')
        if search_directory_by_cod_sub(Input_code) is None:
            print('Документ отсутствует')
        else:
            print('Документ хранится на полке:', search_directory_by_cod_sub(Input_code))
            

# Функция отображени полной информации по всем документам
def full_information():
    if Input_command == 'I':
        for document_in_list in documents:
            directory = search_directory_by_cod_sub(document_in_list['number'])
            print('№: ' +document_in_list['number']+ ', тип: ' +document_in_list['type']+ ', владелец: ' 
                  +document_in_list['name']+', полка хранения:', directory)
            
            
# Функция добавления новой полки
def add_new_directory():
    if Input_command == 'ads':
        Input_new_directory = input('Введите номер полки: ')
        if Input_new_directory in directories:
            print('Такая полка уже существует. Текущий перечень полок: ', ", ".join(list(directories.keys())))
        else:
            directories[Input_new_directory] = []
            print('Полка добавлена. Текущий перечень полок: ', ", ".join(list(directories.keys())))
            
            
# Функция удаления полки, если она пустая
def remove_directory():
    if Input_command == 'ds':
        Input_directory_for_remove = input('Введите номер полки: ')
        if Input_directory_for_remove in directories:
            if directories[Input_directory_for_remove] == []:
                del directories[Input_directory_for_remove]
                print('Полка удалена. Текущий перечень полок: ', ", ".join(list(directories.keys())))
            else:
                print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ', 
                      ", ".join(list(directories.keys())))
        else:
            print('Такой полки не существует. Текущий перечень полок: ', ", ".join(list(directories.keys())))
```

```python
# Код вызова функций по командам
while True:
    Input_command = input('Введите команду: ')
    if Input_command == 'p':
        search_name_by_code()
    elif Input_command == 's':
        search_directory_by_cod()
    elif Input_command == 'I':
        full_information()
    elif Input_command == 'ads':
        add_new_directory()
    elif Input_command == 'ds':
        remove_directory()
    elif Input_command == 'q':
        print('Работа программы завершена')
        break
    else:
        print("Введите корректную команду")
```
