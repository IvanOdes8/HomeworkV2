```python
data = pd.read_csv(r'D:\Netology\Pandas\Lesson_2_basic\power.csv')
data.head()
```

```python
## фильтр стран Прибалтики по категориям 4, 12 и 21 и годам
Baltic = data[ (data.country == "Latvia") | (data.country == "Lithuania") | (data.country == "Estonia")]
Category_of_Baltic = Baltic[ (Baltic.category == 4) | (Baltic.category == 12) |(Baltic.category == 21) ]
Category_of_Baltic_2005_2010 = Category_of_Baltic[ (Category_of_Baltic.year > 2005) & (Category_of_Baltic.year < 2010)]
Category_of_Baltic_2005_2010.head()
```

```python
## сумма quantity>0 для стран Прибалтики по категориям 4, 12 и 21
Summa_of_quantity_Baltic = Category_of_Baltic_2005_2010[ Category_of_Baltic_2005_2010.quantity > 0 ]['quantity'].sum()
Summa_of_quantity_Baltic
```
